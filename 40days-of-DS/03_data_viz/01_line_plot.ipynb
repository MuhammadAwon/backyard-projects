# Import libraries
Seaborn automatically installs these libraries
* numpy
* scipy
* pandas
* matplotlib

```python
# Import libraries
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset
phool = sns.load_dataset("iris")
phool.head()
```

```python
# Draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.show();
```

## Adding titles

```python
# Draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("Phoolon ka Plot")

plt.show();
```

## Adding axes limits

```python
# Draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("Phoolon ka Plot")
plt.xlim(2)
plt.ylim(1)

plt.show();
```

## Set styles
* darkgrid
* whitegrid
* dark
* white
* ticks

```python
# Style must be set before plotting otherwise it will throw "NameError"
sns.set_style(style="dark")

# Draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("Phoolon ka Plot")

plt.show();
```

## Size of figure

```python
# Change figure size
plt.figure(figsize=(10,8))

# Draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("Phoolon ka Plot")

plt.show();
```
