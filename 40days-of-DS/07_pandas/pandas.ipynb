## Import Libraries

```python
import numpy as np
import pandas as pd
```

```python
# Create object
s = pd.Series([1,3,5,np.nan,7,8,4])
s
```

```python
# Create date object
dates = pd.date_range("20220119", periods=7)
dates
```

```python
# Convert dates object into dataframe
df = pd.DataFrame(np.random.randn(7,4), index=dates, columns=list("ABCD"))
df
```

```python
# Create dataframe from dictionary
df2 = pd.DataFrame(
    {
        "A": 1.2,
        "B": pd.Timestamp("20220119"),
        "C": pd.Series(1, index=list(range(4)), dtype="float32"),
        "D": np.array([3] * 4, dtype="int32"),
        "E": pd.Categorical(["test", "train", "test", "train"]),
        "F": "foo"
    }
)

df2
```

```python
# Find data types of all columns in the dataframe
df2.dtypes
```

```python
# View data
df.head(2)
```

```python
df.tail(3)
```

```python
# Find rows in the dataframe
df.index
```

```python
df2.index
```

```python
# Convert dataframe into numpy array
df.to_numpy()
```

```python
# Find statistical values
df.describe()
```

```python
# Transpose dataframe
df.T
```

```python
# Sort elements in the dataframe (row wise)
df.sort_index(axis=0, ascending=False)
```

```python
# Sort dataframe values by a specific column
df.sort_values(by="C")
```

```python
# Selection of dataframe
df["B"]
```

```python
# Row wise slicing
df[2:6]
```

```python
# Selection with labels
df.loc[dates[0]]
```

```python
# Select multiple columns
df.loc[:, ["B", "C"]]
```

```python
# Slice through rows with specific columns
df.loc["20220120":"20220124", ["A", "B", "C"]]
```

```python
# Extract value from specific row
df.loc["20220124", ["A", "B", "C"]]
```

```python
# Slicing through iloc
df.iloc[3:10]
```

```python
# iloc slicing on specify columns
df.iloc[:, 0:2]
```

```python
# Boolean operations on dataframe
df[df["B"] > 0.2]
```

```python
df[df > 0]
```

```python
# Copy dataframe
df3 = df.copy()
```

```python
# Add new column in df3
df3["E"] = ["one", "two", "three", "four", "five", "six", "seven"]
df3
```

```python
# Update df3 (remove column E)
df3 = df3.loc[:, :"D"]
df3
```

```python
# Calculate row wise mean
df3["mean"] = df3.mean(axis=1)
df3
```
