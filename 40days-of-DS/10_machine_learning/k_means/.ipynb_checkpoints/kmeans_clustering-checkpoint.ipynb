# K-Means Clustering

The k-means clustering is a simple algorithm capable of clustering the unlabeled data very quickly and efficiently.

Let's see a practical implementation of that. We'll be using **iris** dataset for this purpose.

```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
phool = sns.load_dataset('iris')
phool.head()
```

We remove Target Feature `species` and only deal with Despcriptive features (i.e. `sepal_length, sepal_width, petal_length, petal_width`).

```python
X = phool.iloc[:,:-1].values
X[:5] # first five samples
```

```python
# Frequency distribution of species
phool_frequency = pd.crosstab(index=phool['species'], columns='count')
phool_frequency
```

## Finding the optimum number of clusters for k-means classification

```python
# Elbow method
from sklearn.cluster import KMeans  
  
# Using for loop for iterations from 1 to 10  
elbow_list= []  
for i in range(1, 11):  
    kmeans = KMeans(n_clusters=i, random_state= 42)  
    kmeans.fit(X)  
    elbow_list.append(kmeans.inertia_)  

# Plot the results onto a line graph to observe "Elbow"
plt.plot(range(1, 11), elbow_list)  
plt.title('The Elbow Method Graph')  
plt.xlabel('Number of clusters(k)')  
plt.ylabel('With_in_cluster_sum_of_squares(WCSS)')  
plt.show()
```

From the plot above, we can see the elbow point is at 3. So the number of clusters here will be 3.

*elbow point**: sum of squares of the distances of each data point in all clusters to their respective centroids.

Its time to train the k-means algorithm on the data and make predictions.

## Apply K-means Clustering

```python
# Train model
model = KMeans(n_clusters=3, random_state=42)

# Predictions
y_pred = model.fit_predict(X)
y_pred
```

## Visualizing the Clusters

### Plot clusters on the basis of Sepal Features

```python
# For setosa cluster
plt.scatter(X[y_pred==1,0], X[y_pred==1,1], s=100, c='red', label='setosa')

# For versicolor cluster
plt.scatter(X[y_pred==0,0], X[y_pred==0,1], s=100, c='green', label='versicolor')

# For virginica cluster
plt.scatter(X[y_pred==2,0], X[y_pred==2,1], s=100, c='blue', label='virginica')

# For centroid
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],
            s=200, marker='X', c='black', label='Centroid')

# Show plot
plt.title('Clusters of phool species')  
plt.xlabel('sepal_length')  
plt.ylabel('sepal_width')  
plt.legend()  
plt.show()
```

```python
# 3d scatterplot using matplotlib
fig = plt.figure(figsize = (15,15))
ax = fig.add_subplot(111, projection='3d')

plt.scatter(X[y_pred==1,0], X[y_pred==1,1], s=100, c='red', label='setosa')
plt.scatter(X[y_pred==0,0], X[y_pred==0,1], s=100, c='green', label='versicolor')
plt.scatter(X[y_pred==2,0], X[y_pred==2,1], s=100, c='blue', label='virginica')

#Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=200, marker='X', c='black', label='Centroids')
plt.show()
```

### Plot clusters on the basis of petal Features

```python
# For setosa cluster
plt.scatter(X[y_pred==1,2], X[y_pred==1,3], s=100, c='red', label='setosa')

# For versicolor cluster
plt.scatter(X[y_pred==0,2], X[y_pred==0,3], s=100, c='green', label='versicolor')

# For virginica cluster
plt.scatter(X[y_pred==2,2], X[y_pred==2,3], s=100, c='blue', label='virginica')

# For centroid
plt.scatter(kmeans.cluster_centers_[:,2], kmeans.cluster_centers_[:,3],
            s=200, marker='X', c='black', label='Centroid')

# Show plot
plt.title('Clusters of phool species')  
plt.xlabel('petal_length')  
plt.ylabel('petal_width')  
plt.legend()  
plt.show()
```

```python
# 3d scatterplot using matplotlib
fig = plt.figure(figsize = (15,15))
ax = fig.add_subplot(111, projection='3d')

plt.scatter(X[y_pred==1,2], X[y_pred==1,3], s=100, c='red', label='setosa')
plt.scatter(X[y_pred==0,2], X[y_pred==0,3], s=100, c='green', label='versicolor')
plt.scatter(X[y_pred==2,2], X[y_pred==2,3], s=100, c='blue', label='virginica')

#Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:,2], kmeans.cluster_centers_[:,3], s=200, marker='X', c='black', label='Centroids')
plt.show()
```

## Accuracy check with Known Labels i.e Species in Iris Dataset

```python
target = phool.iloc[:,-1:]
target
```

### Convert Categorical into Numeric Data

```python
target = target.replace('setosa', 1)
target = target.replace('versicolor', 0)
target = target.replace('virginica', 2)
target
```

### Accuracy Measure

```python
from sklearn.metrics import accuracy_score
score = accuracy_score(target, y_pred)
print(f'K-means clustering accuracy: {score:.3f}')
```

### Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(target, y_pred)
cm
```

```python
# Plot confusion matrix
sns.set_style(style='whitegrid')
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True)
plt.xlabel('Predicted Output')
plt.ylabel('Actual Output')
plt.title('Confusion Marix')
plt.show();
```

We have 2 `versicolor` and 14 `virginica` predictions misclassified.
