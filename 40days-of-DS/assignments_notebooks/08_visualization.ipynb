```python
# Draw a bar plot (remove error bars)
sns.barplot(x="Location", y="Age (years)-Float/Int",
            hue="Gender", data=stud_df,
           order=["East", "West", "India", "Pakistan"],
           color="blue", ci=None)
plt.show();
```

```python
# Draw a bar plot (use color palette)
sns.barplot(x="Location", y="Age (years)-Float/Int",
            hue="Gender", data=stud_df,
           order=["East", "West", "India", "Pakistan"],
           color="red", ci=None, palette="Paired")
plt.show();
```

```python
# Draw a bar plot (use estimator (mean))
sns.barplot(x="Location", y="Age (years)-Float/Int",
            hue="Gender", data=stud_df,
            estimator=np.mean, palette="Paired")
plt.show();
```

```python
# Draw a bar plot (use estimator (median) and color saturation)
sns.barplot(x="Location", y="Age (years)-Float/Int",
            hue="Gender", data=stud_df,
            estimator=np.median, palette="Paired",
           saturation=1)
plt.show();
```

```python
# Draw a bar plot (horizontal)
sns.barplot(x="Your Weight in kg? (float)", y="Blood group ",
            hue="Gender", data=stud_df,
            estimator=np.median, palette="Paired",
           saturation=0.2)
plt.show();
```

```python
# Draw a bar plot (with custom parameters)
sns.barplot(x="Blood group ", y="Your Weight in kg? (float)", data=stud_df,
           linewidth=3, facecolor=(0.1, 0.2, 0.5, 0.1),
           errcolor=".5", edgecolor="0")
plt.show();
```

### What are the four components of facecolor?
facecolor recognizes the RGBA (red, green, blue, alpha) format to specify colors. It's a tuple of float values ranging from 0 to 1.

## Customze Plots

```python
# View dataset
stud_df.head()
```

```python
# Draw boxplot (gender and age)
sns.boxplot(x="Gender",
           y="Age (years)-Float/Int",
           data=stud_df)
plt.show();
```

```python
# Draw boxplot (show mean)
sns.boxplot(x="Gender",
           y="Age (years)-Float/Int",
           data=stud_df,
           showmeans=True)
plt.show();
```

```python
# Draw boxplot (show mean along with custom values of mean properties)
sns.boxplot(x="Gender",
           y="Age (years)-Float/Int",
           data=stud_df,
           showmeans=True,
           meanprops={"marker": "+",
                      "markersize": "12",
                      "markeredgecolor": "red"})
plt.show();
```

```python
# Draw boxplot (show labels with different sizes, and font style)
sns.boxplot(x="Gender",
           y="Age (years)-Float/Int",
           data=stud_df,
           showmeans=True,
           meanprops={"marker": "+",
                      "markersize": "12",
                      "markeredgecolor": "red"})

# Show labels
plt.xlabel("Gender", size=14)
plt.ylabel("Age (years)", size=12)
plt.title("Students gender and their age", size=16,
          weight="bold") # weight use for font style

plt.show();
```

### What is facet wrap and facet grid?
* **facet wrap** - wraps a 1d ribbon of panels into 2d.

* **facet grid** - takes a dataframe as input and the names of the variables that will form the row, column or hue dimensions of the grid. The variables (columns) should be categorical and the data at each level of the variable will be used for a facet along that axis.

```python
# View dataset
stud_df.head()
```

```python
# Draw multi-plot grid (facet grid) for plotting conditional relationships

# Construct the grid
g = sns.FacetGrid(stud_df,
              col="Where do you live?",
              row="Gender",
              height=3.5)

# Map the facegrid construct to draw the plot
g.map(sns.scatterplot,
      "Your Weight in kg? (float)",
     "Age (years)-Float/Int");
```
